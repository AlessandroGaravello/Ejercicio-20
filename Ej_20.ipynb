```python
import pandas as pd
import matplotlib.pyplot as plt
datos = pd.read_csv("products.csv")
df = datos[["unitsInStock", "categoryID"]]


df.head(25).plot.barh(x="unitsInStock",y="categoryID")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
datos = pd.read_csv("products.csv")
df = datos[["unitPrice"]]


df.unitPrice.value_counts().head(15).plot.pie()
plt.show()
```
